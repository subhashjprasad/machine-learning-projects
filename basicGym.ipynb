Importing Libraries & Defining Functions to Display Video

```python
# imports
import gym
from gym.wrappers import RecordVideo
import glob
import io
import base64
from IPython.display import HTML
from pyvirtualdisplay import Display
from IPython import display as ipythondisplay

# ignoring deprecation warnings
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

# function to display video using complicated mp4 encoding stuff
def show_video():
    mp4list = glob.glob('video/*.mp4')
    if len(mp4list) > 0:
        for videoNum in range(1):
            mp4 = mp4list[videoNum]
            video = io.open(mp4, 'r+b').read()
            encoded = base64.b64encode(video)
            ipythondisplay.display(HTML(data='''<video alt="test" autoplay
                loop controls style="height: 400px;">
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
                </video>'''.format(encoded.decode('ascii'))))
    else:
        print("Could not find video")

# wrapping the environment in the RecordVideo class to make the video showable
def wrap_env(env):
    env = RecordVideo(env, './video', episode_trigger = lambda episode_number: True)
    return env
```

Trying Out MountainCarV0 Environment

```python
env = wrap_env(gym.make("MountainCar-v0")) # loading the MountainCarv0 environment
observation = env.reset() # resetting environment and getting initial state

for i in range(1000):
    # place for the agent
    action = env.action_space.sample() # getting a random action
    observation, reward, done, info = env.step(action) # take one step and record state

    if done: # iteration has ended
        observation = env.reset()

env.close()
show_video()
```
